```python
# import packages
import requests
from bs4 import BeautifulSoup
```

```python
response = requests.get('https://www.bookswagon.com/')
response
```

```python
html_data = response.text
```

```python
data = BeautifulSoup(html_data, 'html.parser')
print(data.prettify())
```

```python
## Print Link for all books titles

l1,l2,l3=[],[],[]
all_books_titles = data.find_all(class_='booktitle')
for i in all_books_titles:
    l1.append(i.string)
print(l1)

all_books_authors = data.find_all(class_='author')
for j in all_books_authors:
    l2.append(j.string)
print(l2)

all_books_price = data.find_all(class_='actualprice')
for k in all_books_price:
    l3.append(k.string)
print(l3)

print(len(l1))
print(len(l2))
print(len(l3))
l3 = l3[:-1]
print(len(l3))
```

```python
import pandas as pd

dictionary = {'title':l1,'author':l2,'price':l3}
dataset = pd.DataFrame(dictionary)
dataset.head()
```

```python
dataset.to_csv('books.csv')
```
